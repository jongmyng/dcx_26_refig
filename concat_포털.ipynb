```python
import pandas as pd
import glob
import os
import unicodedata
```

```python
FOLDER_PATH = "."
PORTAL_KEYWORDS = ["ë„¤ì´ë²„í¬íƒˆ"]
```

```python
# ëª¨ë“  CSV íŒŒì¼ ê°€ì ¸ì˜¤ê¸°
all_csv = glob.glob(os.path.join(FOLDER_PATH, "*.csv"))

# í‚¤ì›Œë“œë¡œ í•„í„°ë§ (ìœ ë‹ˆì½”ë“œ ì •ê·œí™” ì ìš©)
files = []
for f in all_csv:
    basename = unicodedata.normalize('NFC', os.path.basename(f))
    if any(kw in basename for kw in PORTAL_KEYWORDS):
        files.append(f)

print(f"ğŸ“ ë°œê²¬ëœ í¬í„¸ íŒŒì¼ ({len(files)}ê°œ):")
for f in sorted(files):
    print(f"   {os.path.basename(f)}")
```

```python
# ë³‘í•©
dfs = []
for file in sorted(files):
    try:
        try:
            df = pd.read_csv(file, encoding='utf-8-sig')
        except:
            df = pd.read_csv(file, encoding='cp949')
        df['_source_file'] = os.path.basename(file)
        dfs.append(df)
        print(f"âœ“ {os.path.basename(file)}: {len(df)}í–‰")
    except Exception as e:
        print(f"âœ— {os.path.basename(file)}: {e}")

if dfs:
    df_portal = pd.concat(dfs, ignore_index=True)
    df_portal.to_csv("combined_í¬í„¸.csv", index=False, encoding='utf-8-sig')
    print(f"\nâœ… ì´ {len(df_portal)}í–‰ â†’ combined_í¬í„¸.csv")
else:
    print("âš ï¸ ë³‘í•©í•  íŒŒì¼ì´ ì—†ìŠµë‹ˆë‹¤.")
```
