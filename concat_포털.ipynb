# 🌐 포털용 CSV 병합

```python
import pandas as pd
import glob
import os
import unicodedata
```

```python
FOLDER_PATH = "."
PORTAL_KEYWORDS = ["네이버포탈"]
```

```python
# 모든 CSV 파일 가져오기
all_csv = glob.glob(os.path.join(FOLDER_PATH, "*.csv"))

# 키워드로 필터링 (유니코드 정규화 적용)
files = []
for f in all_csv:
    basename = unicodedata.normalize('NFC', os.path.basename(f))
    if any(kw in basename for kw in PORTAL_KEYWORDS):
        files.append(f)

print(f"📁 발견된 포털 파일 ({len(files)}개):")
for f in sorted(files):
    print(f"   {os.path.basename(f)}")
```

```python
# 병합
dfs = []
for file in sorted(files):
    try:
        try:
            df = pd.read_csv(file, encoding='utf-8-sig')
        except:
            df = pd.read_csv(file, encoding='cp949')
        df['_source_file'] = os.path.basename(file)
        dfs.append(df)
        print(f"✓ {os.path.basename(file)}: {len(df)}행")
    except Exception as e:
        print(f"✗ {os.path.basename(file)}: {e}")

if dfs:
    df_portal = pd.concat(dfs, ignore_index=True)
    df_portal.to_csv("combined_포털.csv", index=False, encoding='utf-8-sig')
    print(f"\n✅ 총 {len(df_portal)}행 → combined_포털.csv")
else:
    print("⚠️ 병합할 파일이 없습니다.")
```
